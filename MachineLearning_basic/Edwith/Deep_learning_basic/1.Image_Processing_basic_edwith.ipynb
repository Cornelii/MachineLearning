# Image processing practice

```python
import numpy as np
import os
from scipy.misc import imread, imresize
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cwd=os.getcwd()
def print_typeshape(img):
    print("Type is {}".format(type(img)))
    print("Shape is {}".format(img.shape,))
```

```python
amber = imread("./data/image_basic/amber-heard-aquaman.jpg")
print_typeshape(amber)
```

```python
plt.figure(figsize=(13,11))
plt.imshow(amber)
plt.title("amber heard in aquqman")
plt.draw()
```

```python
amber2 = amber.astype(np.float)
```

```python
plt.figure(figsize=(13,11))
plt.imshow(amber2)
plt.title("amber heard in aquqman")
plt.draw()
```

```python
amber2=amber/255
plt.figure(figsize=(13,11))
plt.imshow(amber2)
plt.title("amber heard in aquqman")
plt.draw()
```

# Resize

```python
amber_resized = imresize(amber, [100,100])
print_typeshape(amber_resized)

plt.figure(figsize=(3,5))
plt.imshow(amber_resized)
plt.title("resized amber heard in aquqman")
plt.draw()
```

# GrayScale

```python
def rgb2gray(rgb):
    if len(rgb.shape) is 3:
        return np.dot(rgb,[0.299, 0.587, 0.144])
    else:
        print("Current image is gray!")
        return rgb
```

```python
gray_amber = rgb2gray(amber_resized)
```

```python
print_typeshape(gray_amber)
```

```python
plt.imshow(gray_amber, cmap=plt.get_cmap('gray'))
```

```python
ambervec = np.reshape(gray_amber,(1,-1));
print_typeshape(ambervec)
```

# Load from directory

```python
cwd=os.getcwd()
path = cwd + "/data/image_basic/"
flist = os.listdir(path)
print(flist)
```

```python
for i, f in enumerate(flist):
    print("{}th image: {}".format(i,f))
```

```python
valid_exts=[".jpg",".gif",".png",".bmp",".tga",".jpeg"]
imgs = []
names = []
for f in flist:
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_exts:
        continue
        
    fullpath = os.path.join(path,f)
    imgs.append(imread(fullpath))
    
    names.append(os.path.splitext(f)[0])
    
```

```python
os.path.splitext("amber-heard-aquaman.jpg")
```

# Plot loaded images

```python
for img, name in zip(imgs, names):
    plt.imshow(img)
    plt.title(name)
    plt.show()
```
